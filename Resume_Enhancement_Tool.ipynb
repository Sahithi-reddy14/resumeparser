```python
import re
def cleanResume(resumeText):
    resumeText = re.sub('http\S+\s*', ' ', resumeText)  # remove URLs
    resumeText = re.sub('RT|cc', ' ', resumeText)  # remove RT and cc
    resumeText = re.sub('#\S+', '', resumeText)  # remove hashtags
    resumeText = re.sub('@\S+', '  ', resumeText)  # remove mentions
    resumeText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', resumeText)  # remove punctuations
    resumeText = re.sub(r'[^\x00-\x7f]',r' ', resumeText)
    resumeText = re.sub('\s+', ' ', resumeText)  # remove extra whitespace
    return resumeText

resumeDataSet['cleaned_resume'] = resumeDataSet.Resume.apply(lambda x: cleanResume(x))
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
from collections import defaultdict

#Load BERT model for embeddings
bert_model = SentenceTransformer('paraphrase-MiniLM-L12-v2')

#Define category-specific skills
category_skills = {
    "Java Developer": ["java", "spring boot", "microservices", "hibernate", "multithreading"],
    "Database": ["sql", "database design", "data warehousing", "mongodb", "postgresql"],
    "Data Science": ["machine learning", "data analysis", "deep learning", "python", "tensorflow", "pytorch", "sql"],
    "Advocate": ["legal research", "contract law", "litigation", "intellectual property"],
    "HR": ["recruitment", "employee relations", "performance management", "training and development"],
    "DotNet Developer": ["c#", ".net core", "asp.net", "entity framework"],
    "Hadoop": ["hadoop", "mapreduce", "hive", "pig", "spark"],
    "DevOps Engineer": ["docker", "kubernetes", "ci/cd", "jenkins", "ansible"],
    "Automation Testing": ["selenium", "pytest", "junit", "test automation framework"],
    "Testing": ["manual testing", "regression testing", "performance testing"],
    "Civil Engineer": ["autocad", "structural analysis", "construction management"],
    "Business Analyst": ["business process modeling", "requirement gathering", "stakeholder analysis"],
    "SAP Developer": ["sap abap", "sap hana", "sap fiori"],
    "Health and Fitness": ["nutrition", "exercise science", "personal training"],
    "Python Developer": ["python", "flask", "django", "pandas", "numpy"],
    "Arts": ["painting", "graphic design", "illustration"],
    "Electrical Engineering": ["circuit design", "embedded systems", "power systems"],
    "Sales": ["lead generation", "customer relationship management", "negotiation"],
    "Network Security Engineer": ["firewall management", "penetration testing", "vpn configuration"],
    "Mechanical Engineer": ["cad", "manufacturing processes", "thermal systems"],
    "ETL Developer": ["etl pipelines", "data integration", "informatica", "talend"],
    "Blockchain": ["smart contracts", "ethereum", "solidity", "hyperledger"],
    "Operations Manager": ["supply chain management", "process optimization", "logistics"],
    "Web Designing": ["html", "css", "javascript", "ux/ui design"],
    "PMO": ["project management", "risk assessment", "agile methodology"]
}


def extract_bert_features(text_data):
    #Converts text into numerical representation using BERT embeddings.
    embeddings = bert_model.encode(text_data, convert_to_tensor=True)
    print("BERT Feature Extraction Completed...\n")
    return embeddings.cpu().numpy()

def split_data(features, target_data, test_size=0.2, random_state=42):
    #Splits the data into training and testing sets.
    X_train, X_test, y_train, y_test = train_test_split(
        features, target_data, random_state=random_state, test_size=test_size, stratify=target_data)
    print(f"Training Data Shape: {X_train.shape}")
    print(f"Testing Data Shape: {X_test.shape}\n")
    return X_train, X_test, y_train, y_test

def compute_similarity(skill_list, resume_embeddings):
    #Computes similarity scores for the given list of desired skills.
    skill_vector = bert_model.encode([" ".join(skill_list)], convert_to_tensor=True).cpu().numpy()
    similarity_scores = cosine_similarity(skill_vector, resume_embeddings)
    return similarity_scores

def extract_missing_skills(category, resume_text):
    #Identifies missing skills based on category and resume content.
    if category in category_skills:
        required_skills = category_skills[category]
    else:
        required_skills = []  # Default empty list if category is unknown
    missing_skills = [skill for skill in required_skills if skill.lower() not in resume_text.lower()]
    return missing_skills

def rank_resumes_by_similarity(similarity_scores, resume_data, top_n=5):
    #Ranks resumes based on similarity scores and displays the top matched resumes.
    top_resume_indices = np.argsort(similarity_scores[0])[::-1]  # Sort in descending order
    print(f"Top {top_n} Matched Resumes for Given Skills:\n")

    for i, index in enumerate(top_resume_indices[:top_n]):
        category = resume_data.iloc[index]['Category']
        resume_text = resume_data.iloc[index]['cleaned_resume']
        similarity_score = similarity_scores[0][index]
        missing_skills = extract_missing_skills(category, resume_text)

        print(f"Rank {i+1}: - {category}")
        print(f"Resume Text: {resume_text[:500]}...")  # Print first 500 chars for readability
        print(f"Similarity Score: {similarity_score:.4f}")
        print(f"Missing Skills: {', '.join(missing_skills) if missing_skills else 'None'}\n")


resume_texts = resumeDataSet['cleaned_resume'].tolist()
resume_embeddings = extract_bert_features(resume_texts)

desired_skills = sum(category_skills.values(), [])  # Flatten list of all skills
similarity_scores = compute_similarity(desired_skills, resume_embeddings)

rank_resumes_by_similarity(similarity_scores, resumeDataSet, top_n=5)
```

```python
def print_all_resumes(similarity_scores, resume_data):
    #Prints all resumes ranked by similarity scores.
    all_resume_indices = np.argsort(similarity_scores[0])[::-1]  # Sort in descending order
    print("All Matched Resumes for Given Skills:\n")
    for i, index in enumerate(all_resume_indices):
        category = resume_data.iloc[index]['Category']
        resume_text = resume_data.iloc[index]['cleaned_resume']
        similarity_score = similarity_scores[0][index]
        missing_skills = extract_missing_skills(category, resume_text)

        print(f"Rank {i+1}: - {category}")
        print(f"Resume Text: {resume_text[:500]}...")  # Print first 500 chars for readability
        print(f"Similarity Score: {similarity_score:.4f}")
        print(f"Missing Skills: {', '.join(missing_skills) if missing_skills else 'None'}\n")

print_all_resumes(similarity_scores, resumeDataSet)
```
